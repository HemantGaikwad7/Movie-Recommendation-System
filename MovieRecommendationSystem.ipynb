#Abstract

The digital era has ushered in a transformative era for the entertainment industry, notably in how audiences discover and consume movies. With the burgeoning array of options available, personalized recommendation systems have become essential tools to enhance user engagement by effectively suggesting films that cater to individual tastes. This project presents the development of a movie recommendation system that leverages content-based filtering techniques to provide personalized movie suggestions.

The core of the recommendation system is a Python function, recommend(movie), which utilizes a pre-processed dataset (new_df) containing movie titles and associated metadata. The system computes similarities between movies using a similarity matrix (sim), which is presumed to be pre-calculated using relevant features such as genre, cast, director, and user ratings.

When a user inputs a movie title, the system identifies the index of this movie in the dataset and retrieves a list of movies sorted by their similarity scores in descending order. The top five movies, excluding the input movie itself, are then displayed to the user as recommendations. This approach ensures that the recommendations are dynamically tailored to the user's current selection, enhancing the discovery experience and potentially increasing satisfaction and platform retention.

The implementation of this system demonstrates the practical application of machine learning techniques in real-world scenarios, offering a scalable solution that can be extended to include more complex algorithms and richer metadata for even more precise recommendations.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/HemantGaikwad7/Movie-Recommendation-System/main/MoviesDataset.csv'
movies = pd.read_csv(url)
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['tags'] = movies['genre'] + movies['overview'] + movies['original_language']
movies.head()
```

```python
new_df = movies[['id', 'title','genre', 'overview' ,'tags']]
new_df.head()
```

```python
new_df= new_df.drop(columns=['genre','overview'])
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title'] == 'Avatar']
```

```python
dist = sorted(list(enumerate(sim[0])),reverse=True,key = lambda vec: vec[1])
```

```python
dist
```

```python
for i in dist[0:5]:
  print(new_df.iloc[i[0]].title)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = sorted(list(enumerate(sim[movie_index])),reverse=True,key = lambda vec: vec[1])
  for i in distances[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
